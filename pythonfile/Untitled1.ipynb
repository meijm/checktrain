```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

# 设置 ChromeDriver 路径
chromedriver_path = r'C:\Users\Administrator\Documents\meijmPy\chromedriver.exe'

# 启动 Chrome 浏览器
driver = webdriver.Chrome(chromedriver_path)

# 要查询的车次和日期
train_no = 'G1532'
date = '2023-05-01'

# 打开查询页面
url = 'https://www.12306.cn/index/'
driver.get(url)

# 输入车次和日期，按下回车键
fromStation = driver.find_element_by_id('fromStationText')
toStation = driver.find_element_by_id('toStationText')
trainDate = driver.find_element_by_id('train_date')

fromStation.clear()
fromStation.send_keys('上海')
fromStation.send_keys(Keys.ENTER)

toStation.clear()
toStation.send_keys('北京')
toStation.send_keys(Keys.ENTER)

trainDate.clear()
trainDate.send_keys(date)
trainDate.send_keys(Keys.ENTER)

# 点击“查询”按钮
search_button = driver.find_element_by_id('search_one')
search_button.click()

# 等待页面跳转
time.sleep(10)

# 切换到新页面
driver.switch_to.window(driver.window_handles[-1])

# 获取表格数据
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 获取表格数据

```

```python
table = soup.find( id='queryLeftTable')

# 解析表格数据并存储到 Pandas DataFrame 中
data = []
for tr in table.find_all('tr')[1:]:
    tds = tr.find_all('td')
    print(tr.text)
    station = tds[0].text.strip()
    arrival_time = tds[1].text.strip()
    departure_time = tds[2].text.strip()
    seat_types = [td.text.strip() for td in tds[3:]]
    data.append([station, arrival_time, departure_time] + seat_types)

columns = ['车站', '到站时间', '发车时间', '商务座', '一等座', '二等座', '高级软卧', '软卧', '硬卧', '软座', '硬座', '无座']
df = pd.DataFrame(data, columns=columns)

# 输出查询结果
print(df)

# 关闭浏览器
driver.quit()
```
